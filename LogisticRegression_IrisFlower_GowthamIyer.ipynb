# Creating Logistic Regression model using worst performing input feature (Sepal Width)

```
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

X = dataset.drop(columns='class_of_iris')
X_sepal_width = dataset['sepal_width']
y = dataset['class_of_iris']

label_encoder = preprocessing.LabelEncoder()
y_label_encoder = label_encoder.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X_sepal_width, y_label_encoder,
                                                    test_size=0.3, random_state=0)

X_train = np.reshape(X_train.values, (-1,1))
X_test = np.reshape(X_test.values, (-1,1))

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

# Evaluating models using metrics: accuracy, precision, recall, F1-score.

```
print('Evalutation metrics of worst performing input - Sepal Width')
y_pred = logreg.predict(X_test)
print('Accuracy of test set: {:.2f}'.format(logreg.score(X_test, y_test)))

confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)

report = classification_report(y_test, y_pred)
print(report)
```
